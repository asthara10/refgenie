# <img src="../img/refgenie_logo.svg" class="img-header"> hello world! 

To test `refgenie`, we'll index the human mitochondrial genome. Just run this command:

Without docker: `refgenie -i http://big.databio.org/example_data/rCRS.fa.gz`

With docker: `refgenie -i http://big.databio.org/example_data/rCRS.fa.gz -d`

But first, let's look at the `--help` output to get an idea of usage options:

```bash
refgenie --help
```

Now, let's go ahead and explore the output. This will produce a lot of logging information from all the indexers. Scroll down past all this to get to the next section where we'll explore the output.

```bash
refgenie -i http://big.databio.org/example_data/rCRS.fa.gz -d
```

We were able to index that mitochondrial genome in just a matter of seconds (full genomes will obviously take much longer). Let's see what this process produced:

```bash
tree /ext/yeti/genomes/rCRS
```

Here you can see that we have indexes for a few common indexers. All of the above output will be saved in the `refgenie_log.md` file.
